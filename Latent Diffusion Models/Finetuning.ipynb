```python
from diffusers import DiffusionPipeline, StableDiffusionXLImg2ImgPipeline
import torch

model = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype=torch.float16,
)
pipe.to("cuda")
```

```python
prj_path = "/content/maheshbabu"
pipe.load_lora_weights(prj_path, weight_name="maheshbabu_pytorch_lora_weights.safetensors")
```

```python
prompt = "a rebel maheshbabu person standing in a fighting pose covered in dirt with a runny nose and fists up determined look --ar 1:2 --q 1.25 --s 1000"

seed = 42
generator = torch.Generator("cuda").manual_seed(seed)
image = pipe(prompt=prompt, generator=generator).images[0]
# image = refiner(prompt=prompt, generator=generator, image=image).images[0]
image.save(f"generated_image.png")
```
